## 4. Logical operators
Logical operators are used to combine conditional statements:


#### using 'and (⋂ )' operator <br>
T ⋂ F = False <br>
F ⋂ T = False <br>
T ⋂ T = True <br>
F ⋂ F = True

```python
1 > 2 
```

```python
3 > 4 
```

```python
1 > 2  and 3 > 4 
```

```python
'fifi' == 'fifi'
```

```python
'fifi' == 'durrotun'
```

```python
'fifi' == 'fifi' and 'fifi' == 'durrotun'
```

```python
'fifi' == 'durrotun' and 'fifi' == 'fifi'
```

#### using 'or (⋃ )' operator

T ⋃  F = True <br>
F ⋃  T = True <br>
T ⋃  T = True <br>
F ⋃  F = False

```python
1 == 2
```

```python
4 == 5
```

```python
1 == 2 or 4 == 5
```

```python
'Lima' != 'Enam'
```

```python
'Satu' == 'Dua'
```

```python
'Lima' != 'Enam' or  'Satu' == 'Dua'
```

```python
'Lima' != 'Enam' or  'Satu' == 'Dua' or 4 == 5
```
